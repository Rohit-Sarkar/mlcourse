# Topic 01 : Pandas Data Analysis

```python
import pandas as pd
import numpy as np

df = pd.read_csv('G:/My Data Science files/Topic01/adult_data.csv')
df.head()
```

```python
# to look the shape of DataFrame
print(df.shape)

# to look into the columns
print(df.columns)

#some general information of Dataset
print(df.info())
```

```python
# Change the column type 
# df['salary'] = df['salary'].astype('int64')

# to know the mean and percentile of the dataset
df.describe()
```

```python
#to know the statistics of non numerical features.

df.describe(include=['object'])
```

```python
# to know the distribution of categorical features

df['salary'].value_counts()
```

```python
# to know the proportion of categorical features

df['salary'].value_counts(normalize=True)
```

```python
# Sorting the values.

df.sort_values(by='education-num',ascending=False).head()
```

```python
# Sorting multiple columns 

df.sort_values(by=['education-num', 'age'],
        ascending=[True, False]).head()
```

```python
## Applying the methods

df.apply(np.max)
```

```python
df[df['race'].apply(lambda race: race == 'Black')].head()
```

```python
d = {'Male' : 0, 'Female' : 1}
df['sex'] = df['sex'].map(d)
df.head()

# the same thing can be done by replace method.
# df = df.replace({'sex': d})
```

```python
#Grouping 
columns_to_show = ['age', 'capital-gain', 
                   'hours-per-week']

df.groupby(['salary'])[columns_to_show].describe(percentiles=[])
```

```python
# Cross tabulation

pd.crosstab(df['salary'],df['workclass'])
```

```python
pd.crosstab(df['salary'],df['workclass'],normalize=True)
```

```python
# Pivot Table


df.pivot_table(['capital-gain', 'capital-loss', 'hours-per-week'],
               ['race'], aggfunc='mean')
```

```python
#df.drop(['Capital_Gain_phw'],axis=1,inplace=True)
```

```python
gain_per_hr = (df['capital-gain'] / df['hours-per-week']) 
df.insert(loc=len(df.columns), column='Capital_Gain_phw', value=gain_per_hr) 
# loc parameter is the number of columns after which to insert the Series object
# we set it to len(df.columns) to paste it at the very end of the dataframe
df.head()

### we can create more simply like
df['Capiatl_Gain_phw'] = (df['capital-gain'] / df['hours-per-week']) 
```
